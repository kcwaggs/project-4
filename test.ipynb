```python
import pandas as pd

# Define the speechiness buckets based on the threshold values
def categorize_speechiness(speechiness):
    if speechiness < 0.33:
        return 'non-speech'
    elif speechiness >= 0.33 and speechiness <= 0.66:
        return 'mixed'
    else:
        return 'speech'

# Apply the categorization function to create a new column
maindf_encoded['speechiness_category'] = maindf_encoded['speechiness'].apply(categorize_speechiness)

# One-hot encode the speechiness categories
maindf_encoded = pd.get_dummies(maindf_encoded, columns=['speechiness_category'], prefix='speechiness', prefix_sep='_')
```

```python
# Create a function to find the most popular genre for an artist based on genre counts
def find_most_popular_genre(artist_genres, genre_counts):
    max_count = 0
    most_popular_genre = None
    for genres_list in artist_genres:
        for genre in genres_list:
            if genre in genre_counts and genre_counts[genre] > max_count:
                max_count = genre_counts[genre]
                most_popular_genre = genre
    return most_popular_genre

# Iterate through each song in the filtered_df DataFrame
for index, row in filtered_df.iterrows():
    all_artist_genres = eval(row['artists_genres'])
    most_popular_genre = find_most_popular_genre(all_artist_genres, genre_counts)
    filtered_df.at[index, 'artist_most_popular_genre'] = most_popular_genre

# Display the updated DataFrame with the new column "artist_most_popular_genre"
print(filtered_df)
```
